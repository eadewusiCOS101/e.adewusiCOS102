```python
def get_parameters():
    p = float(input("What  is the principal in naira?"))
    r = float(input("What is the interest rate in percentage?"))
    t = float(input("What is the time in years?"))
    print("")
    return [p , r , t]


print("Welcome to the interest amount calculator\n")
while True:
    process = input("\nWhich interest amount would you like to calculate?\n1)Simple Inerest. \n2)Compound Interest. \n3)Annuity. \nPlease use the number code for each operation to run it.\n\nuse Keyword 'Quit' to stop the program.\n")
    if process == "1":
        parameters = get_parameters()
        p = parameters[0]
        r = parameters[1]
        t = parameters[2]

        a = p * (1 + (r * t / 100))
        print(f"\nThe simple interest amount is ₦{a:.2f}\n")
    elif process == "2":
        parameters = get_parameters()
        p = parameters[0]
        r = parameters[1]
        t = parameters[2]

        a = p * (1 + ((r / 100) ** t))
        print(f"\nThe compound interest amount is ₦{a:.2f}\n")
    elif process == "3":
        parameters = get_parameters()
        p = parameters[0]
        r = parameters[1]
        t = parameters[2]

        a = p * ((((1 + (r / t)) ** t) - 1) / (r / t))
        print(f"\nThe annuity plan amount is ₦{a:.2f}\n")
    elif process.lower() == "quit":
        print("\nShutting down the program.")
        break
    else:
        print("\nIncorrect, choose a process using the number code\nor write quit.")
```
